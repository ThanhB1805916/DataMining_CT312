```python
import json

with open(os.path.join(path, "Src/news_train.json"), "r", encoding="utf-8") as file:
  news_data = json.load(file)
```

```python
topics = news_data.keys()
```

```python
for n in news_data:
  print(f"{n}:len:{len(news_data[n])}")
```

```python
# để cân bằng dữ liệu nên để len = min
news_trim = {}

for n in news_data:
  news_trim[n] = news_data[n][:1000]
```

```python
len(news_trim)
```

```python
import pandas as pd

def dict_to_df(dict):
  df = pd.DataFrame()

  for t in dict:
    _df = pd.DataFrame(dict[t], columns=["News"])
    _df["Topics"] = t

    df = df.append(_df)

  return df
```

```python
df = dict_to_df(news_trim)
```

```python
df.to_pickle(os.path.join(path, "Src/news_df.pkl"))
```

```python
import pickle 

with open(os.path.join(path, "Src/topics.pkl"), "wb") as wb:
  pickle.dump(list(topics), wb)
```

```python
def store_df(data, line):
  news = {}

  for n in data:
    news[n] = data[n][:line]

  df = dict_to_df(news)
  df.to_pickle(os.path.join(path, f"Src/news_df_{line}.pkl"))
```

```python
store_df(news_trim, 10)
store_df(news_trim, 50)
store_df(news_trim, 100)
store_df(news_trim, 500)
```
