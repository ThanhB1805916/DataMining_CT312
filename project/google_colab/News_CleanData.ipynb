```python
import re
import string

# xóa các ký tự đặc biệt chỉ chừa lại chữ
def clean_text(text):

    # viết thường
    text = text.lower()

    # xóa dấu ngoặc
    text = re.sub('[\(\[].*?[\)\]]', '', text)

    # dấu nháy
    text = re.sub("[''‘’“”…]", '', text)

    # bỏ dấu câu (, . : ;)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)

    # xóa từ có số
    text = re.sub('\w*\d\w*', '', text)

    # xóa dấu new line
    text = re.sub('\n', ' ', text)

    return text
```

```python
clean_df = pd.DataFrame(df.News.apply(lambda x: clean_text(x)))
```

```python
clean_df["Topics"] = df["Topics"]
```

```python
import pickle
with open(os.path.join(path, "Src/stop_words.pkl"), 'rb') as rb:
   vn_sw = pickle.load(rb)
```

```python
# gom stop word sang từ việt
vn_stop_words = []

for w in vn_sw:
  vn_stop_words.append(ViTokenizer.tokenize(w))
```

```python
# chắc có tiếng anh nên gom thêm stopwords
from sklearn.feature_extraction import text

stop_words = text.ENGLISH_STOP_WORDS.union(vn_stop_words)
```

```python
len(stop_words)
```

```python
with open(os.path.join(path, "Clean/stop_words.pkl"), 'wb') as wb:
    pickle.dump(stop_words, wb)
```
