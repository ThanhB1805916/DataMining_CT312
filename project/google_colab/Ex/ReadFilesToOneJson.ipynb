```python
import os
from google.colab import output 

def get_news(folder_path):
    news = {}

    # duyện qua các thư mục
    dirs = os.listdir(folder_path)
    for file_name in dirs:
        file_paths = os.listdir(os.path.join(folder_path, file_name))

        topics = []

        # lấy file trong từ thư mục
        for file_path in file_paths:
            with open(os.path.join(folder_path, file_name, file_path), 'r', encoding="utf-16") as file:
                text = file.read()
                
                topics.append(text)

        news[file_name] = topics

    return news
```

```python
news_train_dict = get_news("/content/Train_Full")
```

```python
len(news_train_dict)
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/Project/Data/"
```

```python
import json

with open(os.path.join(path, "news_train.json"), 'w', encoding='utf-8') as file:
 json.dump(news_train_dict, file, ensure_ascii=False)
```
